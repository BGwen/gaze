```python
import numpy as np
from scipy.misc import face
from scipy.ndimage import zoom
from scipy.special import logsumexp
import torch
from PIL import Image
import matplotlib.pyplot as plt
from pysaliency.plotting import visualize_distribution

import deepgaze2e

DEVICE = 'cpu'

# you can use DeepGazeI or DeepGazeIIE
model = deepgaze2e.DeepGazeIIE(pretrained=True).to(DEVICE)

# Define a list of image file paths
image_files = ["37-11-1k.jpg", "Picture1.jpg"]
#["1-11.jpg", "11-11.jpg", "15-11.jpg", "2-11.jpg", "20-11.jpg",
               #"28-11.jpg", "3-11.jpg", "37-11.jpg", "43-11.jpg", "6-11.jpg"]  # Use the same image for demonstration purposes

# Create subplots
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(12, 30))

# Iterate over each image file
for i, image_file in enumerate(image_files):
    # Load image
    img = Image.open(image_file)
    image = np.array(img) 

    # Create centerbias template
    centerbias_template = np.zeros((256, 256))

    # Rescale centerbias to match image size
    centerbias = zoom(centerbias_template, (image.shape[0]/centerbias_template.shape[0], 
                                              image.shape[1]/centerbias_template.shape[1]), 
                      order=0, mode='nearest')

    # Renormalize log density
    centerbias -= logsumexp(centerbias)

    # Convert image to tensor
    image_tensor = torch.tensor([image.transpose(2, 0, 1)]).to(DEVICE)
    centerbias_tensor = torch.tensor([centerbias]).to(DEVICE)

    # Predict log density of fixations
    log_density_prediction = model(image_tensor, centerbias_tensor)

    # Plot image and fixations
    axs[i, 0].imshow(image)
    axs[i, 0].set_axis_off()

    axs[i, 1].matshow(log_density_prediction.detach().cpu().numpy()[0, 0])  # first image in batch, first (and only) channel
    axs[i, 1].set_axis_off()

    visualize_distribution(log_density_prediction.detach().cpu().numpy()[0, 0], ax=axs[i, 2])
    axs[i, 2].set_axis_off()

# Show all images with fixations
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import zoom
from scipy.special import logsumexp
import torch
from PIL import Image

import deepgaze_pytorch

DEVICE = 'cpu'

# Load DeepGazeIII model
model = deepgaze_pytorch.DeepGazeIII(pretrained=True).to(DEVICE)

# Define a list of image file paths
image_files = ["1-11.jpg", "11-11.jpg", "15-11.jpg", "2-11.jpg", "20-11.jpg"]#,
               #"28-11.jpg", "3-11.jpg", "37-11.jpg", "43-11.jpg", "6-11.jpg"]

# Create subplots
fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(16, 20))

# Flatten the axs array to iterate over subplots
axs = axs.flatten()

# Initialize lists to store all fixation coordinates
all_fixations_x = []
all_fixations_y = []

# Iterate over each image file
for i, image_file in enumerate(image_files):
    # Load image
    img = Image.open(image_file)
    image = np.array(img) 

    # Create centerbias template
    centerbias_template = np.zeros((1024, 1024))

    # Rescale centerbias to match image size
    centerbias = zoom(centerbias_template, (image.shape[0]/centerbias_template.shape[0], 
                                              image.shape[1]/centerbias_template.shape[1]), 
                      order=0, mode='nearest')

    # Renormalize log density
    centerbias -= logsumexp(centerbias)

    # Convert image to tensor
    image_tensor = torch.tensor([image.transpose(2, 0, 1)]).to(DEVICE)
    centerbias_tensor = torch.tensor([centerbias]).to(DEVICE)

    # Define fixation history function
    def get_fixation_history(fixation_coordinates, model):
        history = []
        for index in model.included_fixations:
            try:
                history.append(fixation_coordinates[index])
            except IndexError:
                history.append(np.nan)
        return history

    # Initialize random state
    rst = np.random.RandomState(seed=23)

    # Generate starting point for the scanpath
    center_x = image.shape[1] // 2
    center_y = image.shape[0] // 2
    start_x = np.random.randint(center_x - 5, center_x + 5)
    start_y = np.random.randint(center_y - 5, center_y + 5)

    # Initialize fixations with the starting point
    fixations_x = [start_x]
    fixations_y = [start_y]

    # Generate fixations for current image
    for j in range(9):  # Keep this loop fixed at 9 iterations as the starting point is already included
        # Get fixation history
        x_hist = get_fixation_history(fixations_x, model)
        y_hist = get_fixation_history(fixations_y, model)

        # Convert to PyTorch tensors
        x_hist_tensor = torch.tensor([x_hist]).to(DEVICE)
        y_hist_tensor = torch.tensor([y_hist]).to(DEVICE)

        # Predict log density of fixations
        log_density_prediction = model(image_tensor, centerbias_tensor, x_hist_tensor, y_hist_tensor)
        logD = log_density_prediction.detach().cpu().numpy()[0, 0]

        # Sample next fixation coordinates
        next_x, next_y = sample_from_logdensity(logD, rst=rst)

        # Append next fixation coordinates to sequence
        fixations_x.append(next_x)
        fixations_y.append(next_y)

    # Plot image and fixations
    # Plot image and fixations
    axs[i*2].imshow(image)
    axs[i*2].plot(fixations_x, fixations_y, 'o-', color='red')
    axs[i*2].scatter(fixations_x[0], fixations_y[0], 80, color='blue', zorder=100)  # First fixation in blue
    axs[i*2].scatter(fixations_x[-1], fixations_y[-1], 80, color='yellow', zorder=100)  # Last fixation in yellow
    axs[i*2].set_axis_off()

    axs[i*2+1].matshow(log_density_prediction.detach().cpu().numpy()[0, 0])  # first image in batch, first (and only) channel
    axs[i*2+1].plot(fixations_x, fixations_y, 'o-', color='red')
    axs[i*2+1].scatter(fixations_x[0], fixations_y[0], 80, color='blue', zorder=100)  # First fixation in blue
    axs[i*2+1].scatter(fixations_x[-1], fixations_y[-1], 80, color='yellow', zorder=100)  # Last fixation in yellow
    axs[i*2+1].set_axis_off()
    
    # Adding all fixation coordinates to the lists
    all_fixations_x.append(fixations_x.copy())
    all_fixations_y.append(fixations_y.copy())

# Show all images with fixations
plt.tight_layout()
plt.show()
```

```python
for idx, fixations_x in enumerate(all_fixations_x):
    print(f"Subplot {idx + 1} fixations_x: {fixations_x}")

for idy, fixations_y in enumerate(all_fixations_y):
    print(f"Subplot {idy + 1} fixations_y: {fixations_y}")
```
